圖片

```python
###解析圖片
import re

def parse_result_file(result_file):
    predictions = {}
    with open(result_file, 'r') as f:
        lines = f.readlines()
        image_path = None
        for line in lines:
            if "Predicted in" in line:
                image_path = line.split(':')[0].strip().split('/')[-1]
            if "container" in line and "left_x" in line:
                match = re.match(r'\s*(\w+): (\d+)%\s*\(left_x:\s*(\d+)\s+top_y:\s*(\d+)\s+width:\s*(\d+)\s+height:\s*(\d+)\)', line)
                if match:
                    left_x = int(match.group(3).strip())
                    top_y = int(match.group(4).strip())
                    width = int(match.group(5).strip())
                    height = int(match.group(6).strip())

                    if image_path not in predictions:
                        predictions[image_path] = []

                    predictions[image_path].append((left_x, top_y, width, height))

                    print(f"{image_path}, {left_x}, {top_y}, {width}, {height}")
    return predictions

# 測試解析函數
result_file = "/content/drive/MyDrive/result_image_accuracy.txt"  # 修改為result.txt路徑
predictions = parse_result_file(result_file)
print(predictions)
```

```python
from PIL import Image
import os

def crop_objects(predictions, images_dir, output_dir):
    # 如果輸出目錄不存在，則創建它
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for image_name, objects in predictions.items():
        # 構建圖片路徑
        image_path = os.path.join(images_dir, image_name)
        # 打開圖片
        with Image.open(image_path) as img:
            # 從圖片名稱中提取擴展名
            image_ext = os.path.splitext(image_name)[1]
            for i, obj in enumerate(objects):
                left_x, top_y, width, height = obj
                # 裁剪物件
                cropped_obj = img.crop((left_x, top_y, left_x + width, top_y + height))
                # 儲存裁剪後的圖片，保持原始圖片的檔名
                output_path = os.path.join(output_dir, f"{image_name}")
                cropped_obj.save(output_path)
                print(f"裁剪後的物件 {i+1} 已儲存到: {output_path}")
```

```python
# 測試裁剪函數
images_dir = "/content/drive/MyDrive/image_accuracy_data"  # 測試圖片目錄
output_dir = "/content/drive/MyDrive/cropped_images"  # 儲存裁剪後物件的目錄
crop_objects(predictions, images_dir, output_dir)
```

影片

```python
###解析影片
import re

def parse_result_video_file(result_file):
    predictions = []
    with open(result_file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if "container" in line and "left_x" in line:
                match = re.match(r'\s*(\w+): (\d+)%\s*\(left_x:\s*(\d+)\s+top_y:\s*(\d+)\s+width:\s*(\d+)\s+height:\s*(\d+)\)', line)
                if match:
                    left_x = int(match.group(3).strip())
                    top_y = int(match.group(4).strip())
                    width = int(match.group(5).strip())
                    height = int(match.group(6).strip())

                    predictions.append((left_x, top_y, width, height))

                    print(f"{left_x}, {top_y}, {width}, {height}")
    return predictions

# 測試解析函數
result_video = "/content/drive/MyDrive/darknet/result_video_10.txt"
predictions_video = parse_result_video_file(result_video)
print(predictions_video)
```

```python
###切割影片
import os
import cv2

# 載入影像文件夾中的所有圖像
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename))
        if img is not None:
            images.append(img)
    return images
# 根據預測值切割圖像
def crop_images(images, predictions):
    cropped_images = []
    for img, prediction in zip(images, predictions):
        left_x, top_y, width, height = prediction
        cropped_img = img[top_y:top_y+height, left_x:left_x+width]
        cropped_images.append(cropped_img)
    return cropped_images

# 影像資料夾路徑
image_folder = "/content/drive/MyDrive/video_10"

# 讀取影像資料夾中的所有圖像
images = load_images_from_folder(image_folder)

# 使用預測值切割圖像
cropped_images = crop_images(images, predictions_video)

# 保存切割後的圖像
output_folder = "/content/drive/MyDrive/cropped_video_10"
os.makedirs(output_folder, exist_ok=True)
for i, cropped_img in enumerate(cropped_images):
    cv2.imwrite(os.path.join(output_folder, f"cropped_image_{i}.jpg"), cropped_img)
```
