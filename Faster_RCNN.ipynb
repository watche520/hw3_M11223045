```python
import os
import pandas as pd
import cv2
from google.colab.patches import cv2_imshow

# 定義路徑
input_dir = '/content/OCR_DataSet'
output_dir = '/content/spilt'
weights_path = '/content/outputs/training/custom_training/best_model.pth'
inference_script = '/content/fastercnn-pytorch-training-pipeline/inference.py'

# 設定要測試的圖片數量
num_images_to_test = 35  # 這裡設置你想測試的圖片數量
processed_images = 0

# 確保輸出目錄存在
os.makedirs(output_dir, exist_ok=True)

## 遍歷輸入目錄中的所有圖片
for filename in os.listdir(input_dir):
    if filename.endswith('.jpg'):
        input_image_path = os.path.join(input_dir, filename)

        # 進行推論
        inferred_image_name = 'inferred_' + filename
        inferred_image_dir = os.path.join(output_dir, inferred_image_name)
        os.makedirs(inferred_image_dir, exist_ok=True)
        os.system(f'python {inference_script} --input {input_image_path} --weights {weights_path} --output {inferred_image_dir} --table')

        # 動態構建推論結果的 CSV 檔案路徑
        csv_path = os.path.join(inferred_image_dir, 'boxes.csv')  # 假設推論結果的 CSV 檔名為 boxes.csv
        if not os.path.exists(csv_path):
            print(f"CSV file not found for {filename}, skipping.")
            continue

        # 讀取推論結果的 CSV 檔案
        df = pd.read_csv(csv_path, sep=' ')

        # 讀取原始圖片
        image = cv2.imread(input_image_path)

        # 獲取切割區域並保存
        for index, row in df.iterrows():
            xmin = int(row['xmin'])
            xmax = int(row['xmax'])
            ymin = int(row['ymin'])
            ymax = int(row['ymax'])

            # 切割圖片
            cropped_image = image[ymin:ymax, xmin:xmax]

            # 保存切割後的圖片
            cropped_image_path = os.path.join(inferred_image_dir, f'cropped_{filename}_{index}.jpg')
            cv2.imwrite(cropped_image_path, cropped_image)

            # 顯示切割後的圖片（可選）
            # cv2_imshow(cropped_image)

        # 保存原始圖片
        original_image_path = os.path.join(inferred_image_dir, 'original_' + filename)
        cv2.imwrite(original_image_path, image)

        # 保存標註後的圖片（假設推論後的圖片已保存）
        annotated_image_path = os.path.join(inferred_image_dir, 'annotated_' + filename)
        # 這裡假設標註後的圖片已由推論腳本保存

        # 保存標註的 CSV 資料
        annotated_csv_path = os.path.join(inferred_image_dir, f'annotated_{filename}.csv')
        df.to_csv(annotated_csv_path, index=False)

        # 增加已處理圖片的數量
        processed_images += 1

        # 檢查是否達到測試圖片數量
        if processed_images >= num_images_to_test:
            break

print("指定數量的圖片處理完成。")
```
